```python
m = folium.Map(location=[37.483300,127.036233], zoom_start=12)
for i in range(len(smoke_place)):
  if '개방형' in smoke_place.loc[i, '구분']:
    color = 'red'
    tooltip = '(개방형)' + smoke_place.loc[i, '위치']
  else:
    color = 'blue'
    tooltip = '(폐쇄형)' + smoke_place.loc[i, '위치']
  folium.Marker(location=[smoke_place.loc[i,'위도'], smoke_place.loc[i, '경도']], tooltip=tooltip,icon=folium.Icon(color=color)).add_to(m)
m
```
