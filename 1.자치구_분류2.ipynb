```python
def sil_score(x_data, n_clu=5, metrics='euclidean', linkage='ward', KM=False, DB=False, HC=False, MS=False, GMM=False):

  if KM:
    km_s = []
    for i in range(3, n_clu):
      k = KMeans(n_clusters=i)
      k = k.fit_predict(x_data)
      k_s = silhouette_score(x_data, k, metric=metrics)
      km_s.append(k_s)
    return km_s

  if HC:
    hc_s = []
    for i in range(3, n_clu):
      h = AgglomerativeClustering(n_clusters=i, affinity=metrics, linkage=linkage)
      h = h.fit_predict(x_data)
      h_s = silhouette_score(x_data, h, metric=metrics)
      hc_s.append(h_s)
    return hc_s

  if GMM:
    gmm_s = []
    for i in range(3, n_clu):
      g = GaussianMixture(n_components=n_clu)
      g = g.fit_predict(x_data)
      g_s = silhouette_score(x_data, g, metric=metrics)
      gmm_s.append(g_s)
    return gmm_s

  # if MS:
  #   ms_s = []
  #   for i in range(10):
  #     bw = estimate_bandwidth(x_data, quantile=quantile)
  #     m = MeanShift(bandwidth=bw)
  #     ms = m.fit_predict(X)
  #     m_s = silhouette_score(x_data, ms, metric=metrics)
  #     centers = m.cluster_centers_
  #     uniq_y_ms = np.unique(ms) 
  #     ms_s.append(m_s)
  #     quantile += 0.01
  #   return ms_s, len(uniq_y_ms)
```
