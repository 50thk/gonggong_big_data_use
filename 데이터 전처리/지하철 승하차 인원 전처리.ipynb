```python
#날짜별 지하철 승하차 정보 불러오기
sub01 = pd.read_csv('Data/CARD_SUBWAY_MONTH_202101.csv',index_col = False)
sub02 = pd.read_csv('Data/CARD_SUBWAY_MONTH_202102.csv',index_col = False)
sub03 = pd.read_csv('Data/CARD_SUBWAY_MONTH_202103.csv',index_col = False)
sub04 = pd.read_csv('Data/CARD_SUBWAY_MONTH_202104.csv',index_col = False)
sub05 = pd.read_csv('Data/CARD_SUBWAY_MONTH_202105.csv',index_col = False)
sub06 = pd.read_csv('Data/CARD_SUBWAY_MONTH_202106.csv',index_col = False)
sub07 = pd.read_csv('Data/CARD_SUBWAY_MONTH_202107.csv',index_col = False)
sub08 = pd.read_csv('Data/CARD_SUBWAY_MONTH_202108.csv',index_col = False)

```

```python
#1년 평균 승하차 인원 구하기


#데이터 합치기
sub_sum = pd.concat([sub01,sub02,sub03,sub04,sub05,sub06,sub07,sub08])
#sub_sum

sub_groub = sub_sum.groupby('역명').mean()

sub_mean = sub_groub[['승차총승객수','하차총승객수']]
sub_mean = sub_mean.reset_index()
sub_mean

sub_mean.to_csv

#괄호와 괄호안의 문자 지우기
sub_mean['역명'] = sub_mean['역명'].astype(str).str.replace(r"\(.*\)","")

```

```python
#구 별 승하차 인원 구하기

subway = pd.read_csv('Data/전국도시철도역사정보표준데이터(작업).csv',encoding='cp949')
subway.head()

#서울이 아닌 역 삭제
subway = subway[subway['역사도로명주소'].str.contains('서울')]


target = subway[subway['역사도로명주소'].str.contains('용산')]

target['역사명']

for n in list(set(target['역사명'])):
  print(n)
```

```python

sub_name = sub_mean[sub_mean['역명']=='서울역']
sub_name



df2 = pd.DataFrame([[sub_name.iloc[0,1], sub_name.iloc[0,2]]], columns=['승차총승객수', '하차총승객수'])
df2

df = pd.DataFrame(columns=['역명','승차총승객수', '하차총승객수'])
for n in target['역사명']:
  sub_name = sub_mean[sub_mean['역명']==n]
  df2 = pd.DataFrame([[sub_name.iloc[0,0],sub_name.iloc[0,1],sub_name.iloc[0,2]]], columns=['역명','승차총승객수', '하차총승객수'])
  df = df.append(df2)

df
```

```python
#구 별 승하차 인원 구하기
#name(입력값)으로 구 넣기

def get_subway_pop(name):
  #주소에 구 이름(name)이 포함된 역사를 target에 저장
  target = subway[subway['역사도로명주소'].str.contains(name)]

  #빈 데이터 프레임 생성
  df = pd.DataFrame(columns=['역명','승차총승객수', '하차총승객수'])

  #각 행(각 역)을 for문으로 돌며 빈 데이터 프레임에 해당 역의 승차,하차 승객수 추가
  for n in list(set(target['역사명'])):
    sub_name = sub_mean[sub_mean['역명']==n]
    df2 = pd.DataFrame([[sub_name.iloc[0,0],sub_name.iloc[0,1],sub_name.iloc[0,2]]], columns=['역명','승차총승객수', '하차총승객수'])
    df = df.append(df2)
  return df
```

```python
#승,하차 인원 더하기
def sum_pop(df):
  Sum = df['승차총승객수'].sum() + df['하차총승객수'].sum()
  #반올림 해서 값 반환
  return round(Sum)



gu_list = ['강남','강동','강서','강북','관악','광진','구로','금천','노원','동대문','도봉','동작','마포','서대문','성동','성북','서초','송파','영등포','용산','양천','은평','종로','중구','중랑']

gu_pop =[]
for gu in gu_list:
  gu_pop.append(sum_pop(get_subway_pop(gu)))

print(gu_pop)

subway_pop = pd.DataFrame({ 'gu_name':gu_list, 'sub_pop':gu_pop })
subway_pop
```
